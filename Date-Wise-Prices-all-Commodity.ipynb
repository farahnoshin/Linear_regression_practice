```python
import numpy as np
import pandas as pd
```

```python
Prices = pd.read_csv('Date-Wise-Prices-all-Commodity.csv')
```

```python
Prices.head()
```

```python
Prices.info()
```

```python
Prices_num = Prices.drop(['State', 'District','Market','Commodity','Variety','Grade'],axis=1)
```

```python
Prices_num
```

```python
Prices_num.describe()
```

```python
Prices_num = (Prices_num - Prices_num.min()) / (Prices_num.max() - Prices_num.min())
```

```python
Prices_num.describe()
```

```python
Prices_num
```

```python
Prices_cat_1 = pd.get_dummies(Prices['State'], dtype=int)
```

```python
Prices_cat_1
```

```python
Prices_cat_2 = pd.get_dummies(Prices['District'], dtype=int)
```

```python
Prices_cat_3 = pd.get_dummies(Prices['Market'], dtype=int)
```

```python
Prices_cat_4 = pd.get_dummies(Prices['Commodity'], dtype=int)
```

```python
Prices_cat_5 = pd.get_dummies(Prices['Variety'], dtype=int)
```

```python
Prices_cat_6 = pd.get_dummies(Prices['Grade'], dtype=int)
```

# Merge Numerical and Categorical

```python
Prices_prepared = pd.concat([Prices_cat_1, Prices_cat_2,Prices_cat_3,Prices_cat_4,Prices_cat_5,Prices_cat_6, Prices_num ], axis=1)
```

```python
Prices_prepared
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(Prices_prepared, test_size=0.2, random_state=42)
```

```python
test_x = test_set.drop(['Min_x0020_Price','Max_x0020_Price','Modal_x0020_Price'], axis=1)
```

```python
test_y = test_set[['Min_x0020_Price','Max_x0020_Price','Modal_x0020_Price']]
```

```python
test_x
```

```python
test_y
```

```python
train_x = train_set.drop(['Min_x0020_Price','Max_x0020_Price','Modal_x0020_Price'], axis=1)
train_y = train_set[['Min_x0020_Price','Max_x0020_Price','Modal_x0020_Price']]
```

```python
train_x.shape, train_y.shape
```

```python
test_x.shape, test_y.shape
```

# Learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
```

```python
model = MLPRegressor()
model.fit(train_x, train_y)
```

# Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
test_y_prediction = model.predict(test_x)
```

```python
test_y_prediction
```

```python
test_x.shape
```

```python
test_y_prediction.shape
```

```python
mean_absolute_error(test_y, test_y_prediction)
```

```python
mean_squared_error(test_y, test_y_prediction)
```
